```python
## wczytuję dane o rezerwacjach w systemie hpg
df_hpg = pd.read_csv('hpg_reserve.csv', sep = ',')
df_hpg
```

```python
# sprawdzam czy występują zduplikowane wiersze
df_hpg[df_hpg.duplicated() == True]
```

```python
# usuwam zduplikowane wiersze
df_hpg.drop_duplicates(inplace = True)
df_hpg
```

```python
# sprawdzam czy występują brakujące dane
pd.isnull(df_hpg).sum()
```

```python
# wczytuję dane o restauracjach z podwójnym ID
df_rel = pd.read_csv('store_id_relation.csv', sep = ',')
df_rel
```

```python
# stosując left join i outer join tworzę docelowy dataframe z danymi z obu systemów
df_air_merged =  pd.merge(df_air, df_rel, how = 'left', on = 'air_store_id')

df_hpg_merged = pd.merge(df_hpg, df_rel, how = 'left', on = 'hpg_store_id')

df_merged = pd.merge(df_air_merged, df_hpg_merged, how = 'outer').drop_duplicates()
df_merged
```

```python
# sprawdzam typy zmiennych w df_merged
df_merged.dtypes
```

```python
# konwertuję zmienne reserve_datetime i visit_datetime do typu data-godzina
df_merged['reserve_datetime'] = pd.to_datetime(df_merged['reserve_datetime'])

df_merged['visit_datetime'] = pd.to_datetime(df_merged['visit_datetime'])
```

```python
df_merged.replace(np.nan, 'none', inplace = True, regex = True)
```

```python
# kilka dodatkowych operacji na dataframe - zmiana nazw i kolejności kolumn, sortowanie danych
df_merged.rename(columns = {'air_store_id':'ID_air', 'hpg_store_id':'ID_hpg'}, inplace = True)

df_merged.sort_values(['ID_air', 'ID_hpg', 'reserve_datetime', 'visit_datetime', 'reserve_visitors'], 
                                  ascending = [True, True, True, True, False], inplace = True, ignore_index = True)

df_merged = df_merged[['ID_air', 'ID_hpg', 'reserve_datetime', 'visit_datetime', 'reserve_visitors']]
df_merged
```

```python
# tworzę kolumnę group_visitors, która wprowadza podział rezerwacji w zależności od liczby osób rezerwujących
labels = ['1 - 2', '3 - 5', '6 - 10', '11 - 20', '21 - 50', '51 - 100']

df_merged['group_visitors'] = pd.cut(df_merged.reserve_visitors, [1, 3, 6, 11, 21, 51, 101], right = False, labels = labels)
df_merged
```

```python
# tworzę kolumnę systems, która informuje o systemach z jakich korzysta restauracja
sys = []

for i in df_merged.index:
    if df_merged.ID_air[i] != 'none' and df_merged.ID_hpg[i] != 'none':
        sys.append('air/hpg')
    elif df_merged.ID_hpg[i] == 'none':
        sys.append('air only')
    else:
        sys.append('hpg only')
        
df_merged['systems'] = sys
```

```python
# wykorzystuję ID restauracji w systemach air i hpg do stworzenia multiindexu
df_merged.set_index(['ID_air','ID_hpg'], inplace = True)
```
